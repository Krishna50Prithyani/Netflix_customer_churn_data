```python
# Load dataset
df = pd.read_csv("netflix_customer_churn.csv") 
```

```python
# Drop NaN rows
df = df.dropna() 
```

```python
# Drop identifier columns (UUID, ID, names, etc.)
if "customerid" in df.columns:
    df = df.drop("customerid", axis=1) 
```

```python
# Encode categorical variables
for col in df.select_dtypes(include=['object']).columns:
    if col != "Churn":  # don't encode target
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col]) 
```

```python
# # Define features and target
x = df.drop("churned", axis=1)  
y = df["churned"] 
```

```python
# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42) 
```

```python
#seting up the model
mod1 = DecisionTreeClassifier(max_depth=5, random_state=42)
#run model
mod1.fit(x_train, y_train)
# run the predictions
pred1 = mod1.predict(x_test)
# #evaluate model performance by classification report
print(classification_report(y_test, pred1)) 
```

Precision - How many of the positives predicted are actually positive
Recall - How many of the actual positives were predicted
F1 Score - Harmonic Mean of the precision and recall scores
Support - Number of true samples in that class( in this case, the actual number of customers who stayed)
Accuracy - Overall percentage of correct predictions: True Posivites + True Negatives /Total Samples

```python
#confusion matrix
conf_mat= confusion_matrix(y_test, pred1)

#cm display 
ConfusionMatrixDisplay(conf_mat, display_labels=['No churned', 'churned']).plot() 
```

```python
#plot the decision tree
#get column names from x
x_cols=x.columns
plt.figure(figsize=(9,4), dpi=150)
plot_tree(mod1, fontsize=10, feature_names=x_cols); 
```

```python
#Find which feature is the most important in determining churn
feat_imp= mod1.feature_importances_
plt.figure(figsize= (12,8))
bars = plt.barh(y=x_cols, width=feat_imp);
plt.bar_label(bars, fmt='%1.2f')
plt.xticks(rotation=90)
plt.title("Feature Importance of churn model")
plt.show() 
```

```python
#balancing the dataset
#step 1: Split the dataset into 2 classes:
#no churn df:
df_major=df[df['churned']==0]
#churn df:
df_minor = df[df["churned"]==1] 
```

```python
df_major.shape 
```

```python
df_minor.shape 
```

```python
#Creating down sampled majority data set
df_major_s= df_major.sample(n=len(df_minor),replace=True,random_state=20)
df_major_s.shape 
```

```python
#Merging downsampled majority with minority df
df_ds = pd.concat([df_minor, df_major_s])
df_ds 
```

```python
df_ds= df_ds.reset_index(drop=True) 
```

```python
#Shuffle the data
df_sf= df_ds.sample(frac=1, random_state=41)
df_sf 
```

```python
x= df_sf.drop(["churned","customer_id"], axis=1)
x = pd.get_dummies(x, dtype='int')
y= df_sf["churned"] 
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.20, random_state = 41, stratify = y) 
```

```python
mod2 = DecisionTreeClassifier(max_depth=4)

#run model
mod2.fit(x_train, y_train)
#run predictions
pred2=mod2.predict(x_test)

#evaluate model performance by classification report
print(classification_report(y_test, pred2)) 
```

```python
#confusion matrix
conf_mat= confusion_matrix(y_test, pred2)

#cm display 
ConfusionMatrixDisplay(conf_mat, display_labels=['No churn', 'churned']).plot(); 
```

```python
from skopt import BayesSearchCV
from skopt.space import Real, Categorical, Integer 
```

```python
#defining the search space
param_space = {
    'max_depth': Integer(1, 20), 'criterion':("gini","entropy","log_loss")} 
```

```python
#Initialising the classifier
dt_classifier = DecisionTreeClassifier(random_state = 41) 
```

```python
#Setting the search space
opt = BayesSearchCV(
    dt_classifier,
    param_space,
    n_iter=50,
    cv=3,
    n_jobs=3
) 
```

```python
#Fitting the search to the data
opt.fit(x,y) 
```

```python
#creating a variable with the best parameters
best_param=opt.best_params_
best_param 
```

```python
#creating a new model with the best parameters
mod1b = DecisionTreeClassifier(**best_param) 
```

```python
# fitting the new model with x_train and y_train
mod1b.fit(x_train, y_train)

# running the predictions
pred1b = mod1b.predict(x_test)

# printing the classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, pred1b)) 
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report 
```

```python
#setting and intialising the bayesian model tuning
param_space = {
'n_neighbors':(1, 50),
'weights':("uniform","distance"),
'metric':("minkowski","euclidean","manhattan","chebyshev","hamming")}

kn_classifier= KNeighborsClassifier()

opt3 = BayesSearchCV(
    kn_classifier, 
    param_space,
    n_iter = 32,
    cv=3,
    n_jobs=3)

opt3.fit(x,y) 
```

```python
best_param3 = opt3.best_params_
print(best_param3) 
mod3b = KNeighborsClassifier(**best_param3) 
mod3b.fit(x_train, y_train) 
pred_knc_b = mod3b.predict(x_test) 
print(classification_report(y_test, pred_knc_b)) 
```
